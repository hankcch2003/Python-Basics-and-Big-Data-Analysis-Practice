```python
# 匯入scipy.stats模組，用於進行統計分佈與機率相關的計算
import scipy.stats as stats

# 使用stats.randint()函式建立一個均勻分佈的離散整數型隨機變數rv，範圍為1到10(含1、不含11)
rv = stats.randint(low = 1, high = 11)
```

```python
# 使用rv.rvs()函式從隨機變數rv中抽取5個樣本，並傳回一個包含5個隨機整數的一維陣列
rv.rvs(size = 5)
```

```python
# 使用rv.pmf()函式計算隨機變數rv在點5、7、9、11的機率質量函數值，並傳回一個包含這些點對應機率的一維陣列(若不在定義域內，機率為0)
rv.pmf([5, 7, 9, 11])
```

```python
# 使用rv.cdf()函式計算隨機變數rv在點1、2、3、4、5的累積分佈函數值，並傳回一個包含這些點對應累積機率的一維陣列
rv.cdf([1, 2, 3, 4, 5])
```

```python
# 使用rv.ppf()函式計算隨機變數rv在累積機率0.2、0.5、0.7對應的百分位數值，並傳回一個包含這些百分位數的一維陣列
rv.ppf([0.2, 0.5, 0.7])
```

```python
# 使用rv.median()函式計算並輸出整數區間1到10的均勻離散分佈理論中位數
print(rv.median())
```

```python
# 使用stats.randint.median(1, 11)函式計算並輸出整數區間1到10的均勻離散分佈理論中位數
print(stats.randint.median(1, 11))
```

```python
# 使用rv.mean()函式計算並輸出整數區間1到10的均勻離散分佈理論算術平均
print(rv.mean())
```

```python
# 使用stats.randint.mean(1, 11)函式計算並輸出整數區間1到10的均勻離散分佈理論算術平均
print(stats.randint.mean(1, 11))
```

```python
# 使用rv.var()函式計算並輸出整數區間1到10的均勻離散分佈理論變異數
print(rv.var())
```

```python
# 使用stats.randint.var(1, 11)函式計算並輸出整數區間1到10的均勻離散分佈理論變異數
print(stats.randint.var(1, 11))
```

```python
# 匯入scipy.optimize模組中的root函式，用於求解非線性方程式的根
from scipy.optimize import root

# 定義函式f(x) = 2x² - 4x + 1
def f(x):
    return 2 * x ** 2 - 4 * x + 1

# 使用root()函式從初始值0開始，求解方程式f(x) = 0的根
sol1 = root(f, 0)

# 輸出求得的第一個根
print(sol1.x)

# 使用root()函式從初始值1開始，求解方程式f(x) = 0的根
sol2 = root(f, 1)

# 輸出求得的第二個根
print(sol2.x)

# 使用root()函式從初始值2開始，求解方程式f(x) = 0的根
sol3 = root(f, 2)

# 輸出求得的第三個根
print(sol3.x)
```

```python
# 使用np.roots()函式計算多項式2x² - 4x + 1的所有根
r = np.roots([2, -4, 1])

# 輸出求得的根
print(r)
```

```python
# 定義一個聯立方程式系統
# 2x + y - 5 = 0
# x - 3y + 1 = 0

def fun(x):
    return [
        2 * x[0] + x[1] - 5,
        x[0] - 3 * x[1] + 1
    ]

# 使用root()函式從初始值[0, 0]開始，求解聯立方程式的解
sol = root(fun, [0, 0])

# 輸出求得的解
print(sol.x)
```

```python
# 定義函式f(x) = 2x² - 4x + 1
def f(x):
    return 2 * x ** 2 - 4 * x + 1

# 定義函式g(x) = x - 2
def g(x):
    return x - 2

# 使用root()函式從初始值0開始，求解f(x) - g(x) = 0的根(即求f(x) = g(x)的解)
sol = root(lambda x: f(x) - g(x), 0)

# 輸出求得的根
print(sol.x)

# 輸出根在f(x)的值
print(f(sol.x))

# 輸出根在g(x)的值(理論上應與f(sol.x)相近)
print(g(sol.x))
```

```python
# 匯入scipy.optimize模組中的minimize_scalar函式，用於求解單變量函數的極小值
from scipy.optimize import minimize_scalar

# 定義函式f(x) = 2x² - 4x + 1
def f(x):
    return (2 * x ** 2 - 4 * x + 1)

# 使用minimize_scalar函式求解函式的最小值
res = minimize_scalar(f)

# 輸出結果，並使用format()函式格式化數值
print("當 x 為 {0} 時，函數 f(x) 有最小值為 {1}".format(res.x, f(res.x)))
```

```python
# 匯入scipy.interpolate模組中的interp1d函式，用於一維插值
from scipy.interpolate import interp1d

# 定義函式f(x) = cos(-x² / 9)，稍後用於產生原始數據點
f = lambda x: np.cos(-x ** 2 / 9.0)

# 使用np.linspace()函式建立一個NumPy一維陣列x，從-12到12(包含)，產生25個等距分佈的數值
x = np.linspace(-12, 12, num = 25)

# 計算原始數據對應的y值
y = f(x)

# 建立線性插值函式g1(預設為線性插值，即kind = "linear")
g1 = interp1d(x, y)

# 使用三次樣條插值法建立插值函式g2(kind = "cubic"表示使用三次多項式來逼近資料點)
g2 = interp1d(x, y, kind = "cubic")

# 使用np.linspace()函式建立一個NumPy一維陣列xnew，從-12到12(包含)，產生49個等距分佈的數值
xnew = np.linspace(-12, 12, num = 49)

# 使用線性插值函式g1對新的xnew陣列進行插值，計算對應的ynew1值
ynew1 = g1(xnew)

# 使用三次樣條插值函式g2對新的xnew陣列進行插值，計算對應的ynew2值
ynew2 = g2(xnew)

# 使用plt.plot()函式繪製x對y的點狀圖(使用藍色圓形"bo"標記資料點)，橫軸為x陣列中的元素值，縱軸為對應的函數值，並設定圖例文字為"data"
plt.plot(x, y, "bo", label = "data")

# 使用plt.plot()函式繪製線性插值後的函數曲線(設定為紅色實線"r-")，橫軸為xnew陣列中的元素值，縱軸為插值後的函數值，並設定圖例文字為"linear"
plt.plot(xnew, ynew1, "r-", label = "linear")

# 使用plt.plot()函式繪製三次樣條插值後的函數曲線(設定為綠色虛線"g--")，橫軸為xnew陣列中的元素值，縱軸為插值後的函數值，並設定圖例文字為"cubic"
plt.plot(xnew, ynew2, "g--", label = "cubic")

# 顯示圖例，位置設定為圖表下方中間
plt.legend(loc = "lower center")

# 顯示圖表
plt.show()
```

```python
# 定義一個聯立方程式系統
# x0 * cos(x1) = 4
# x0 * x1 - x1 = 5

def fun(x):
    return [
        x[0] * np.cos(x[1]) - 4,
        x[1] * x[0] - x[1] - 5
    ]

# 使用root()函式從初始值[1, 1]開始，求解聯立方程式的解
sol = root(fun, [1, 1])

# 輸出求得的解
print(sol.x)
```

```python
# 定義函式f(x) = (x - 1) * (x - 2) * (x - 3)，稍後用於產生原始數據點
f = lambda x: (x - 1) * (x - 2) * (x - 3)

# 使用np.linspace()函式建立一個NumPy一維陣列x，從0到5(包含)，產生11個等距分佈的數值
x = np.linspace(0, 5, num = 11)

# 計算原始數據對應的y值
y = f(x)

# 建立線性插值函式g1(預設為線性插值，即kind = "linear")
g1 = interp1d(x, y)

# 使用二次樣條插值法建立插值函式g2(kind = "quadratic"表示使用二次多項式來逼近資料點)
g2 = interp1d(x, y, kind = "quadratic")

# 使用np.linspace()函式建立一個NumPy一維陣列xnew，從0到5(包含)，產生49個等距分佈的數值
xnew = np.linspace(0, 5, num = 49)

# 使用線性插值函式g1對新的xnew陣列進行插值，計算對應的ynew1值
ynew1 = g1(xnew)

# 使用三次樣條插值函式g2對新的xnew陣列進行插值，計算對應的ynew2值
ynew2 = g2(xnew)

# 使用plt.plot()函式繪製x對y的點狀圖(使用藍色圓形"bo"標記資料點)，橫軸為x陣列中的元素值，縱軸為對應的函數值，並設定圖例文字為"data"
plt.plot(x, y, "bo", label = "data")

# 使用plt.plot()函式繪製線性插值後的函數曲線(設定為紅色實線"r-")，橫軸為xnew陣列中的元素值，縱軸為插值後的函數值，並設定圖例文字為"linear"
plt.plot(xnew, ynew1, "r-", label = "linear")

# 使用plt.plot()函式繪製二次樣條插值後的函數曲線(設定為綠色虛線"g--")，橫軸為xnew陣列中的元素值，縱軸為插值後的函數值，並設定圖例文字為"cubic"
plt.plot(xnew, ynew2, "g--", label = "quadratic")

# 顯示圖例，位置設定為圖表上方左邊
plt.legend(loc = "upper left")

# 顯示圖表
plt.show()
```
