```python
# 匯入calendar模組，用來處理與日曆相關的功能
import calendar

# 使用calendar.month函式來取得2024年1月的月曆(以字串格式傳回)
print(calendar.month(2024, 1))
```

```python
# 使用calendar.isleap函式判斷2024年是否為閏年
calendar.isleap(2024)
```

```python
# 匯入calendar模組，並設定別名為cal
import calendar as cal

# 使用cal.month函式顯示2024年1月的月曆(以字串格式輸出)
print(cal.month(2024, 1))
```

```python
# 使用cal.isleap函式判斷2024年是否為閏年
cal.isleap(2024)
```

```python
# 從calendar模組中匯入month函式
from calendar import month

# 顯示2024年1月的月曆(以字串格式輸出)
print(month(2024, 1))
```
