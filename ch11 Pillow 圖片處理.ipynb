# ch11 Pillow 圖片處理

```python
# 從PIL(Python Imaging Library)模組中匯入Image類別，用於處理圖片相關操作
from PIL import Image

# 使用Image.open()函式開啟圖片檔案
# 圖片路徑可以使用雙反斜線(\\)，或是在字串前加上r表示原始字串，以避免跳脫字元錯誤

im = Image.open("E:\\Python\\Python1-基礎Python開發實務\\ch11 圖片存放區\\food.jpg")

# 使用.show()方法顯示圖片，系統會開啟預設的圖片檢視器
im.show()

# 關閉原始圖片檔案，釋放系統資源
im.close()
```

```python
# 輸出圖片的檔案格式，例如JPEG、PNG等
print(im.format)

# 輸出圖片的色彩模式，例如RGB、L(灰階)、RGBA等
print(im.mode)

# 輸出圖片的寬度(單位：像素)
print(im.width)

# 輸出圖片的高度(單位：像素)
print(im.height)

# 輸出圖片的尺寸大小，格式為(寬度, 高度)，單位為像素(pixels)
print(im.size)
```

```python
# 使用Image.open()函式開啟圖片檔案
# 圖片路徑可以使用雙反斜線(\\)，或是在字串前加上r表示原始字串，以避免跳脫字元錯誤

im = Image.open("E:\\Python\\Python1-基礎Python開發實務\\ch11 圖片存放區\\food.jpg")

# 將圖片轉換為1位元黑白模式(非灰階)，只有黑與白兩種顏色
out = im.convert("1")

# 使用.show()方法顯示圖片，系統會開啟預設的圖片檢視器
out.show()

# 將轉換後的圖片存成檔案(命名為food-1.jpg)
out.save("E:\\Python\\Python1-基礎Python開發實務\\ch11 圖片存放區\\food-1.jpg")

# 關閉原始圖片檔案，釋放系統資源
im.close()

# 關閉轉換後的圖片，釋放系統資源
out.close()
```

```python
# 使用Image.open()函式開啟圖片檔案
# 圖片路徑可以使用雙反斜線(\\)，或是在字串前加上r表示原始字串，以避免跳脫字元錯誤

im = Image.open("E:\\Python\\Python1-基礎Python開發實務\\ch11 圖片存放區\\bird.jpg")

# 設定字型與大小(使用Arial字體，字型大小為100)
ttfont = ImageFont.truetype("C:\\Windows\\Fonts\\Arial\\arial.ttf", 100)

# 建立可繪製物件，用來在圖片上繪圖或寫字
draw = ImageDraw.Draw(im)

# 在圖片上指定位置繪製紅色文字"Hello!"，並使用先前設定的字型(Arial字體)
# (150, 10)：文字的繪製位置，左上角座標為x = 150, y = 10
# "Hello!"：要顯示的文字內容
# font = ttfont：使用的字型物件，需事先透過ImageFont.truetype()載入
# fill = (255, 0, 0, 255)：文字的顏色與透明度，這裡是紅色(RGB：255, 0, 0)，完全不透明(Alpha：255)

draw.text((150, 10), "Hello!", font = ttfont, fill = (255, 0, 0, 255))

# 在圖片上指定矩形範圍內繪製紅色橢圓
# [(0, 0), (100, 100)]：定義橢圓所在的矩形框(左上角與右下角座標)
# fill：橢圓的填充顏色與透明度，這裡是綠色(RGB：0, 255, 0)，完全不透明(Alpha：255)

draw.ellipse([(0, 0), (100, 100)], fill = (0, 255, 0, 255))

# 使用.show()方法顯示圖片，系統會開啟預設的圖片檢視器
im.show()

# 將轉換後的圖片存成檔案(命名為bird-6.jpg)
im.save("E:\\Python\\Python1-基礎Python開發實務\\ch11 圖片存放區\\bird-6.jpg")

# 關閉原始圖片檔案，釋放系統資源
im.close()
```
