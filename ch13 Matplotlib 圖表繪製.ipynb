# ch13 Matplotlib 圖表繪製

```python
# 匯入numpy套件，用於數值計算與陣列處理
import numpy as np

# 匯入matplotlib.pyplot模組，用於繪製圖表
import matplotlib.pyplot as plt

# 使用np.arange()函式建立一個NumPy一維陣列，從0開始(包含0)，到小於5結束，間隔值為0.1，元素型別為浮點數
x = np.arange(0, 5, 0.1)

# 使用np.square()函式對陣列x的每個元素取平方，並將結果存放到一維陣列y
y = np.square(x)

# 使用plt.plot()函式繪製x對y的折線圖(預設為藍色實線"-")，橫軸為x陣列中的元素值，縱軸為y陣列中的元素值
plt.plot(x, y)

# 顯示圖表
plt.show()
```
